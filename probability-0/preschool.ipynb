```python
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

SAMPLE = 500000
N_BINS = 59
g0 = np.random.randn(SAMPLE)
g1 = np.random.randn(SAMPLE)
plus = g0 + g1
sub = g0 - g1
mul = g0 * g1

fig, axes = plt.subplots(nrows=1, ncols=1, figsize={10, 4})
ax0 = axes

n, bins, patches = ax0.hist([g0, plus], N_BINS, histtype='bar', normed=1)
ax0.plot(bins, mlab.normpdf(bins, 0, 1), linewidth=4, alpha=0.2, color='#00BEFF')
ax0.plot(bins, mlab.normpdf(bins, 0, math.sqrt(2)), linewidth=4, alpha=0.2, color='#D4CA3A')
ax0.legend(['Normal(0, 1)', 'Normal(0, sqrt(2))', 'X samples', 'Z samples'])
ax0.set_title('X / X + Y')

"""
ax1.hist([g0, mul], N_BINS, histtype='bar')
ax1.set_title('multiple')
"""

fig.tight_layout()
plt.show()
```
