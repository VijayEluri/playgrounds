```python
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

SAMPLE = 500000
N_BINS = 59
g0 = np.random.randn(SAMPLE)
g1 = np.random.randn(SAMPLE)
g2 = np.random.randn(SAMPLE)
g3 = np.random.randn(SAMPLE)
p1 = g0
p2 = g0 + g1
p3 = g0 + g1 + g2
p4 = g0 + g1 + g2 + g3

fig, axes = plt.subplots(nrows=4, ncols=1, figsize={10, 6}, sharex=True, sharey=True)
axs = axes.flatten()[:]
ps = [p1, p2, p3, p4]
sigma = [1, 2, 3, 4]
titles = ['X', 'X + Y', 'X + Y + Z', 'W + X + Y + Z']
for i in range(len(sigma)):
    n, bins, patches = axs[i].hist(ps[i], N_BINS, histtype='bar', width=0.05, normed=1)
    axs[i].plot(bins, mlab.normpdf(bins, 0, math.sqrt(sigma[i])), linewidth=4, alpha=0.3)
    axs[i].set_xlabel(titles[i], fontsize=15)

fig.tight_layout()
plt.show()
```
